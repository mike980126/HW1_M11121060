```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv("adult.data")
df
```

```python
df.apply(lambda x:np.sum(x==' ?'))
```

```python
mask1 = df["workclass"] != " ?"
mask2 = df["native-country"] != " ?"
mask3 = df["occupation"] != " ?"

df2 = df[(mask1& mask2 & mask3 )]
df2
```

```python
df2.apply(lambda x:np.sum(x==' ?'))

#確認df2各欄位無問號資料
```

```python
from sklearn.preprocessing import LabelEncoder
from seaborn import load_dataset

labelencoder = LabelEncoder()
data_le = pd.DataFrame(df2)
data_le = LabelEncoder()

for col in df2[['workclass','education','marital-status','occupation','relationship','race','sex','capital-loss','hours-per-week','native-country','Class']]:
    df2[col] = data_le.fit_transform(df2[col])
df2
```

```python
from sklearn.preprocessing import MinMaxScaler

y = df2['Class'].values
df2_1 = df2.drop(labels=['Class','fnlwgt','education','race','relationship'],axis=1)

df2_1v = df2_1.values

scaler = MinMaxScaler(feature_range=(0,1)).fit(df2_1)
df3 = scaler.transform(df2_1)


print('資料集 dbdpre 的平均值:',df2_1v.mean(axis=0))
print('資料集 dbdpre 的標準差:',df2_1v.std(axis=0))

print('標準化後 資料集 dbdpre 的平均值', df3.mean(axis=0))
print('標準化後 資料集 dbdpre 的標準差', df3.std(axis=0))

#引用來源 https://www.youtube.com/watch?v=P42GqxCXkY8
```

```python
x=df3
x
```

```python
y 
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
from sklearn import tree
import pydot
```

```python
#區分訓練資料及測試資料比例
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)

#trainning Decision Tree 未設alpha時
clf = tree.DecisionTreeClassifier()
clf = clf.fit(x_train, y_train)
```

```python
a = clf.score(x_train, y_train)
print(a)
b = clf.score(x_test, y_test)
print(b)
```

```python
print(classification_report(y_test, clf.predict(x_test)))
```

```python
#未修剪時之決策樹
tree.plot_tree(clf)
```

```python
clean= DecisionTreeClassifier(random_state=0)
path = clean.cost_complexity_pruning_path(x_train, y_train)
ccp_alphas,impurities = path.ccp_alphas,path.impurities
```

```python
#最大有效 alpha 值被移除，因為它是只有一個節點的平凡樹。
fig, ax = plt.subplots()
ax.plot(ccp_alphas[:-1], impurities[:-1], marker='o', drawstyle="steps-post")
ax.set_xlabel("effective alpha")
ax.set_ylabel("total impurity of leaves")
ax.set_title("Total Impurity vs effective alpha for training set")
#引用來源 https://scikit-learn.org/stable/auto_examples/tree/plot_cost_complexity_pruning.html#sphx-glr-auto-examples-tree-plot-cost-complexity-pruning-py
```

```python
clfs = []
for ccp_alpha in ccp_alphas:
    clf = DecisionTreeClassifier(random_state=0, ccp_alpha=ccp_alpha)
    clf.fit(x_train, y_train)
    clfs.append(clf)
print(
    "Number of nodes in the last tree is: {} with ccp_alpha: {}".format(
        clfs[-1].tree_.node_count, ccp_alphas[-1]
    )
)
#引用來源 https://scikit-learn.org/stable/auto_examples/tree/plot_cost_complexity_pruning.html#sphx-glr-auto-examples-tree-plot-cost-complexity-pruning-py
```

```python
clfs = clfs[:-1]
ccp_alphas = ccp_alphas[:-1]

node_counts = [clf.tree_.node_count for clf in clfs]
depth = [clf.tree_.max_depth for clf in clfs]
fig, ax = plt.subplots(2, 1)
ax[0].plot(ccp_alphas, node_counts, marker="o", drawstyle="steps-post")
ax[0].set_xlabel("alpha")
ax[0].set_ylabel("number of nodes")
ax[0].set_title("Number of nodes vs alpha")
ax[1].plot(ccp_alphas, depth, marker="o", drawstyle="steps-post")
ax[1].set_xlabel("alpha")
ax[1].set_ylabel("depth of tree")
ax[1].set_title("Depth vs alpha")
fig.tight_layout()

#節點數量和樹深度隨著 alpha 的增加而減少。
#引用來源 https://scikit-learn.org/stable/auto_examples/tree/plot_cost_complexity_pruning.html#sphx-glr-auto-examples-tree-plot-cost-complexity-pruning-py
```

```python
train_scores = [clf.score(x_train, y_train) for clf in clfs]
test_scores = [clf.score(x_test, y_test) for clf in clfs]

fig, ax = plt.subplots()
ax.set_xlabel("alpha")
ax.set_ylabel("accuracy")
ax.set_title("Accuracy vs alpha for training and testing sets")
ax.plot(ccp_alphas, train_scores, marker="o", label="train", drawstyle="steps-post")
ax.plot(ccp_alphas, test_scores, marker="o", label="test", drawstyle="steps-post")
ax.legend()
plt.show()
#引用來源 https://scikit-learn.org/stable/auto_examples/tree/plot_cost_complexity_pruning.html#sphx-glr-auto-examples-tree-plot-cost-complexity-pruning-py
```
